# Performer Musical Maturity in the Lens of Computation
## Main Question
How do expert and advanced-student level musical performances differ in their use of expressive timing, dynamics, and structural alignment through computational quantitative analysis?

## Hypotheses
- Experts have controlled timing deviations while students have reactive deviations.
- Expert-level performances align dynamics with structure while dynamics in student-level performances emphasize locally. 
- Expert-level performances align deviation in timing and dynamics with structure more consistently than student-level performances.
